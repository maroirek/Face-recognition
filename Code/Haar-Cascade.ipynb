```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
# defining mask  to use 
mask= np.array([[1,1,1],[0,0,0]])
image=cv.imread('5.jpg')
img = cv.cvtColor(image, cv.COLOR_RGB2GRAY)
imgfil= cv.filter2D(img, -1 ,mask)
```

```python
plt.imshow(imgfil, cmap='gray')
```

```python
cv.imshow('With Mask',imgfil)
cv.waitKey(0)
cv.destroyAllWindows()
```
