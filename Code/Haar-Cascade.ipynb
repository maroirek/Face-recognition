```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
# defining mask  to use 
#sobel for edge detection

def Sobel_filter (path) :
    image=cv.imread(path)
    img = cv.cvtColor(image, cv.COLOR_RGB2GRAY)
    kernelxs = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
    kernelys = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
    sobelx = cv.filter2D(img, -1, kernelxs)
    sobely = cv.filter2D(img, -1, kernelys)

    amp_sobel_carre=np.square(sobelx.astype(np.float32))+np.square(sobely.astype(np.float32))
    amp_sobel=np.sqrt(amp_sobel_carre)
    amp_sobel=np.uint8(amp_sobel)
    
    return amp_sobel

plt.imshow(Sobel_filter('5.jpg'), cmap='gray')
```

```python

def Haar_Edges_Masks (path) :
    image=cv.imread(path)
    img = cv.cvtColor(image, cv.COLOR_RGB2GRAY)
    kernelxs = np.array([[1,1,1],[0,0,0]])
    kernelys = np.array([[1,0],[1,0],[1,0]])
    kernelxys = np.array([[1,1,1],[1,1,0],[1,0,0]])
    kernelyxs = np.array([[1,1,1],[0,1,1],[0,0,1]])
    
    Edgex = cv.filter2D(img, -1, kernelxs)
    Edgey = cv.filter2D(img, -1, kernelys)
    Edgexy = cv.filter2D(img, -1, kernelxys)
    Edgeyx = cv.filter2D(img, -1, kernelyxs)

    amp_Edge_carre=np.square(Edgex.astype(np.float32))+np.square(Edgey.astype(np.float32))+np.square(Edgexy.astype(np.float32))+np.square(Edgeyx.astype(np.float32))
    amp_Edge= np.sqrt(amp_Edge_carre)
    amp_Edge= np.uint8(amp_Edge)
    
    return amp_Edge

plt.imshow(Haar_Edges_Masks('5.jpg'), cmap='gray')
```

```python
def Haar_Lines_Masks (path) :
    image=cv.imread(path)
    img = cv.cvtColor(image, cv.COLOR_RGB2GRAY)
    kernelx = np.array([[1,1,1],[0,0,0],[1,1,1]])
    kernely = np.array([[1,0,1],[1,0,1],[1,0,1]])
    kernelxy = np.array([[0,1,1],[1,0,1],[1,1,0]])
    kernelyx = np.array([[1,1,0],[1,0,1],[0,1,1]])
    
    Linex = cv.filter2D(img, -1, kernelx)
    Liney = cv.filter2D(img, -1, kernely)
    Linexy = cv.filter2D(img, -1, kernelxy)
    Lineyx = cv.filter2D(img, -1, kernelyx)

    amp_Line_carre=np.square(Linex.astype(np.float32))+np.square(Liney.astype(np.float32))+np.square(Linexy.astype(np.float32))+np.square(Lineyx.astype(np.float32))
    amp_Line=np.sqrt(amp_Line_carre)
    amp_Line=np.uint8(amp_Line)
    
    return amp_Line

plt.imshow(Haar_Lines_Masks('5.jpg'), cmap='gray')
```
